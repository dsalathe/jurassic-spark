## Group by crops 

Description : 

Grouping crops by category or type could be interesting to look for trends. Indeed there are a lot a different Items in crops which can make the vizualization difficult. 

On the website : http://www.fao.org/faostat/en/#data/TP when we download the file `Item Group` in `Definitions and standards - Crops and livestock products` we observe all the item codes for the different crops. The Item Group Code is the main category which contains many different crops. However those categories look a lot unbalanced, with the number 1882 - Agricult.Products, Total containing most of the crops and the vegetables. Also if we look at quinoa this is what we got :

```python
#Quickly load the data : 
df_group_crops = pd.read_csv('data/GroupCrops.csv')
df_group_crops.shape
```

```python
#Look at the item Quinoa :
df_group_crops[df_group_crops.Item=='Quinoa']
```

We see that the quinoa which has always the same Item Code 92, belongs to different Item Group Code ( Agricult.Products, Total, Cereals , ...) so using the Item Group Code to properly classify the group in which the quinoa belongs is not correct. If we add the contribution of all main group then the total will be bigger than the real production if the same can be found in multiple group also. 

However, if we look at the HS12 Code, its value is unique. The HS12 which is a 

http://unstats.un.org/unsd/tradekb/Attachment439.aspx?AttachmentType=1 (will download the dataset which explain the group)

We have the folowing group : `100850`

If we take the first two numbers (10) then we will have the main category (Cereals). If we take the four first numbers then we go more specific (BuckWheat, Millet, and canary seeds; other cereals). For example, other groups from the main categories are : Oats, Rice, Maize (Corn), Wheat and Meslin, Barley and Rye. 

```python
# Load the UN Comtrade Commodity Classifications :
df_UNCCC = pd.read_excel('data/UNCCC.xlsx')
```

```python
df_HS12 = df_UNCCC[df_UNCCC.Classification=='H4'] # H4 correspond to the HS12
```

```python
# df group crops with the nan values removed
df_group_crops_cleaned = df_group_crops[~df_group_crops['HS12 Code'].isnull()]
```

```python
# If the items is assigned mutliple labels, we only keep the first one
HS12_Code = df_group_crops_cleaned['HS12 Code'].str.split(', ',expand=True).loc[:,0]
```

```python
df_group_crops_cleaned = pd.concat([df_group_crops_cleaned,HS12_Code],axis=1).drop(['Factor','HS Code','HS07 Code','CPC Code','HS12 Code'],axis=1)
```

```python
df_group_crops_cleaned.rename(columns={0:'HS12 Code'},inplace=True)
```

```python
df_group_crops_cleaned['Parent Group'] = df_group_crops_cleaned['HS12 Code'].str[:2].astype(int)
df_group_crops_cleaned['Child Group'] = df_group_crops_cleaned['HS12 Code'].str[:4].astype(int)
```

```python
df_group_crops_cleaned
```

```python
# Check that the columns Parent Group and Child Group are of type int or float
df_group_crops_cleaned.dtypes
```

```python
# Number of parent groups 
df_group_crops_cleaned['Parent Group'].nunique() 
```

```python
df_group_crops_cleaned = df_group_crops_cleaned.drop(['Item Group Code','Item Group'],axis=1).groupby(['Item','Item Code','HS12 Code','Parent Group','Child Group']).count().reset_index()
```

```python
#Merge dummy df with tmp
df_merged_dummies = pd.concat([tmp, df_dummy],axis=1)
df_merged_dummies.head()
```

```python
df_merged_dummies.loc[df_merged_dummies['Import Quantity']==1,'Import Quantity'] = df_merged_dummies.loc[df_merged_dummies['Import Quantity']==1,'Value']
df_merged_dummies.loc[df_merged_dummies['Import Value']==1,'Import Value'] = df_merged_dummies.loc[df_merged_dummies['Import Value']==1,'Value']
df_merged_dummies.loc[df_merged_dummies['Export Quantity']==1,'Export Quantity'] = df_merged_dummies.loc[df_merged_dummies['Export Quantity']==1,'Value']
df_merged_dummies.loc[df_merged_dummies['Export Value']==1,'Export Value'] = df_merged_dummies.loc[df_merged_dummies['Export Value']==1,'Value']
```

```python
list_features_to_drop = ['Item Code','Element Code','Element','Value','Flag','Domain Code','Domain','Item_y']
```

```python
df_merged_dummies.drop(list_features_to_drop,axis=1).groupby(['Area','Item_x','Year'])
```
