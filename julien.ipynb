We notice that in the trade_items dataframe the items are more numerous and different than in the crops. The items are various commodities, some manufactured products and some raw products from animal and agricultural exploitation. 

As we are interested in the crops, our goal is to keep all the items, manufactured and raw, that come from an agricultural product.

We thus make an assumption: the first word of the items in the prod_items list is significant enough to illustrate the product. We will find those keywords, and use them to find the raw products linked to the manufactured one in the trade dataframe.

```
import re
core_words = []
item_to_core = {}
for item in prod_items:
  #Clean the word, to lower case, with no special characters
  splitted_item = re.sub('[^A-Za-z0-9 ]+', '', item.lower()).split()
  core = ''
  ind = 0
  #While the 'core' word is already taken or if it is empty
  while core in core_words or core == '':
    if ind > 0 :
      core += ' '
    # We build the core word by taking the first words in the item, as we observed that the first word is significant enough  
    core += splitted_item[ind]
    ind += 1
  core_words.append(core)
  item_to_core[item] = core
```

MAP TO TRADES

We now want to apply what we did to extract the key words from the trade items.

```
#We will check if the core word is in the item's name. 
#We then sort the core_words list to have the multiple words key words being checked first.
core_words.sort(key=lambda w : len(w), reverse=True)
```

```
#Sad words that did not found their match in the key words
unwanted = []

trade_item_to_core = {}
for item in trade_items:
  cleaned_item = re.sub('[^A-Za-z0-9 ]+', '', item.lower())
  for word in core_words:
    if word in cleaned_item:
      trade_item_to_core[item] = word
      break

  if trade_item_to_core.get(item, '') == '':
    trade_item_to_core[item] = 'NAC' #NAC = Not a crop
    unwanted.append(cleaned_item)
```

```
crops_trade['core_item'] = crops_trade.item.map(lambda item: trade_item_to_core[item])
```

By studying the unwanted list, we can see some items that could be *wanted*. For example, the word *potato* is unwanted, while *potatoes* is a core word. We could use a stemmer to save these types of unwanted words.

```
import nltk

p = nltk.PorterStemmer()

item_to_stemmed_core = {}
for item in prod_items:
  splitted_item = re.sub('[^A-Za-z0-9 ]+', '', item.lower()\
                         .replace(',', ' ').replace('  ', ' '))\
                         .split()
  stem = ''
  ind = 0
  while stem in item_to_stemmed_core.values() or stem == '':
    if ind > 0 :
      stem += ' '
    stem += p.stem(splitted_item[ind])
    ind += 1
  item_to_stemmed_core[item] = stem
item_to_stemmed_core
```

```
item_to_stemmed_core['Sweet potatoes'] = 'sweet potato' #'sweet' was the core word of Sweet potatoes which can be misunderstood (ex: sweet corn)
item_to_stemmed_core['Sugar beet'] = 'sugar beet' #'sugar' was the core word of Sugar beets and sugar can also come from sugar cane.
item_to_stemmed_core['Cow peas, dry'] = 'cow pea' #'cow' was the core word of Cow peas, dry and could then be matched with cow milk produces.

item_to_stemmed_core['Chillies and peppers, dry'] = 'pepper dri' #precision is needed 
item_to_stemmed_core['Chillies and peppers, green'] = 'pepper green'#in this case

item_to_stemmed_core['Onions, shallots, green'] = 'onion' #We switch the 2 values
item_to_stemmed_core['Onions, dry'] = 'onion dri'         #as dry is more specific than multiple types of onions

item_to_stemmed_core['Vegetables Primary'] = 'veget' #Primary becomes the general stopword because there are more infos on it
item_to_stemmed_core['Vegetables, fresh nes'] = 'veget fresh'

item_to_stemmed_core['Chick peas'] = 'chick pea' #added pea as 'chick' can be the key word for 'chicken'
```

```
stemmed_core_words = list(item_to_stemmed_core.values())
stemmed_core_words.sort(key=lambda w : len(w), reverse=True)
unwanted2 = []
trade_item_to_stem = {}

for item in trade_items:
  cleaned_item = [p.stem(word) for word in re.sub('[^A-Za-z0-9 ]+', '', item\
                                              .lower())\
                                              .replace('+', ' ')\
                                              .replace(',', ' ')\
                                              .replace('  ', ' ')\
                                              .split()]
  cleaned_item = ' '.join(cleaned_item)
  for word in stemmed_core_words:
    if word in cleaned_item:
        trade_item_to_stem[item] = word
        break

  if trade_item_to_stem.get(item, '') == '':
    unwanted2.append(item)
trade_item_to_stem
```
